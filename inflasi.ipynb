```python
import pandas as pd

data = pd.read_csv("trending.csv")
data.head()
```

```python
# Tampilkan Kolom

data.columns
```

```python
#buat data baru berdasarkan kolom yang kita inginkan saja

data= data[['title','channel_name', 'category_id','view','like', 'comment','dislike','trending_time',]]
```

```python
data.head()
```

ubah kategory_id jadi category_name . dengan cara import category_names terlebih dahulu

```python
import json

with open("category.json", "rb") as D:
    categories = json.load(D)
```

```python
categories = pd.json_normalize(categories, record_path="items")
categories_dict = {idx:title for idx, title in zip(categories['id'], categories['snippet.title'])}
```

```python
data['category'] = [categories_dict[str(x)]if str(x) in list(categories_dict.keys()) else"Uknown" for x in data['category_id']]
```

cek jika ada nilai bernilai NA

```python
data.isnull().sum()
```

mengubah semua nilai bernilai NA menjadi 0

```python
data = data.fillna(0)
```

mencari nilai tranding_date dan menghapus tranding_time

```python
trending = pd.to_datetime(data['trending_time'])
data['trending_date'] = trending.dt.date
data = data.drop(['trending_time'], axis=1)
```

```python
used_data = data.drop_duplicates(['title','channel_name'], keep='first')
```
